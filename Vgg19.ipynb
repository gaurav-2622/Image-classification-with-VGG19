```python
from tensorflow.keras.applications import VGG19
```

```python
model = VGG19(
    include_top=True,
    weights='imagenet',
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000,
    classifier_activation='softmax'
)
```

```python
model.summary()
```

```python
model.save("vgg19.h5")
```
